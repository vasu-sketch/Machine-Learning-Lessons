# Gradient Descent


## 1) Minimizing a function of a single variable

Given a function $f: \mathbb R \rightarrow \mathbb R$, such as $f(w) = w^2 - 4w + 5$, we want to find 

$$\min_w f(w)$$ 

and the value of $w$ where that minimum is achieved. 

Recall the gradient descent method where we start with a random value $w_0$ and we update $w$ by using the formula

$$
w_{n+1} = w_n - \alpha f'(w_n) 
$$

The value of learning rate $\alpha$ is typically found by experimentation, though more sophisticated techniques are available, such as *line-search* and *trust-region* methods. 

We stop the iterations when the derivative becomes sufficiently close to 0.


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the single variable function f(w)
def f(w):
    return w**2 -4*w + 5

# Define the domain for plotting f(w)
domain = np.linspace(-2, 6, 50)

# Plot the function f(w)
plt.figure(figsize = (8, 6))
plt.plot(domain, f(domain), label = "f(w) = w^2 - 4w+5")
plt.xlabel("w", fontsize = 15)
plt.ylabel("f(w)", fontsize = 15)
plt.legend(fontsize = 15, loc = "upper center");
```

```python
# Start with a guess $w_0=5$

plt.figure(figsize = (8,6))
plt.plot(domain, f(domain), label = "$f(w) = w^2 -4w + 5$")
plt.scatter(5, f(5), color = "magenta", label = "$w_0$")
plt.xlabel("w", fontsize = 15)
plt.ylabel("f(w)", fontsize = 15)
plt.legend(fontsize = 15, loc = "upper center");
```

```python
# Define the derivative 
def df(w): 
    return 2*w - 4

# Choose the random guess
w_0 = 5.0

# Define tangent line function for visualization
def tangent_line(w_i, function, derivative, i = 0, color = "magenta", show = True):
    # Define tangent line
    # y = m*(x - x1) + y1
    def line(w):
        return derivative(w_i)*(w - w_i) + function(w_i)
    wrange = np.linspace(w_i - 1, w_i + 1, 10)
    if show:
        plt.plot(wrange,
            line(wrange), 
            '--', 
            linewidth = 1, 
            color = "red")
    plt.scatter([w_i], [function(w_i)], color = color, label = f"$w_{i}$")


plt.figure(figsize = (8, 6))

plt.plot(domain, f(domain), label = "$f(w) = w^2 - 4w + 5$")

tangent_line(w_0, f, df)

print("w0 =" , w_0)

plt.xlabel("w", fontsize = 15)
plt.ylabel("f(w)", fontsize = 15)
plt.legend(fontsize = 15);
```

```python
# Initialize choice of w
w_0 = 5.0

# Set learning rate 
alpha = 0.8

# Move in the opposite direction of the derivative at w_0
w_1 = w_0 - alpha*df(w_0)

# Print the sequence of w variables
print(f"{w_0 = }")
print(f"{w_1 = }")

# Plot the figure
plt.figure(figsize = (8, 6))
plt.plot(domain, f(domain), label = "$f(w) = w^2 - 4w +5$")

# Visualize the tangent lines
tangent_line(w_0, f, df, show = False)
tangent_line(w_1, f, df, i = 1, color = "green")

# Plot labels
plt.xlabel("w", fontsize = 15)
plt.ylabel("f(w)", fontsize = 15)
plt.legend(fontsize = 15, loc = "upper center");
```

```python
w_2 = w_1 - alpha*df(w_1)

# Print the sequence of w variables
print(f"{w_0 = }")
print(f"{w_1 = }")
print(f"{w_2 = }")

# Plot the figure
plt.figure(figsize = (8, 6))
plt.plot(domain, f(domain), label = "$f(w) = w^2 - 4w + 5$")

# Visualize the tangent lines
tangent_line(w_0, f, df, show = False)
tangent_line(w_1, f, df, i = 1, color = "green", show = False)
tangent_line(w_2, f, df, i = 2, color = "red")

# Plot labels
plt.xlabel("w", fontsize = 15)
plt.ylabel("f(w)", fontsize = 15)
plt.legend(fontsize = 15, loc = "upper center");
```

```python
w_3 = w_2 - alpha*df(w_2)

# Print the sequence of w variables
print(f"{w_0 = }")
print(f"{w_1 = }")
print(f"{w_2 = }")
print(f"{w_3 = }")

# Plot the figure
plt.figure(figsize = (8, 6))
plt.plot(domain, f(domain), label = "$f(w) = w^2 - 4w + 5$")

# Visualize the tangent lines
tangent_line(w_0, f, df, show = False)
tangent_line(w_1, f, df, i = 1, color = "green", show = False)
tangent_line(w_2, f, df, i = 2, color = "red", show = False)
tangent_line(w_3, f, df, i = 3, color = "purple")

# Plot labels
plt.xlabel("w", fontsize = 15)
plt.ylabel("f(w)", fontsize = 15)
plt.legend(fontsize = 15, loc = "upper center");
```

```python
# Compute until you reach 1000 iterations or the derivative is less than 0.001

def derivative_descent(derivative, alpha = 0.8, w_0 = 5.0, max_iter = 1_000):
    W = [w_0]
    i = 0
    while abs(derivative(W[-1])) > 0.001 and i < max_iter:
        w_new = W[-1] - alpha*df(W[-1])
        W.append(w_new)
        i += 1
    W = np.array(W)
    return W

W = derivative_descent(df)

for i, w in enumerate(W):
    print(f"w_{i} = {w} ||| f'(w_{i}) = {df(w)}")
print("")

plt.figure(figsize = (8, 6))
plt.plot(domain, f(domain), label = "$f(w) = w^2 - 4w + 5$")
plt.scatter(W, f(W), color = "magenta")
plt.xlabel("$w$", fontsize = 15)
plt.ylabel("$f(w)$", fontsize = 15)
plt.legend(fontsize = 15, loc = "upper center")
tangent_line(W[18], f, df, i = 18, color = "black");
```

```python
# Consider several choices for the learning rate alpha

alphas = [0.05, 0.5, 0.8, 1.0]

# Call the subplots method for plotting a grid of figures
fig, axs = plt.subplots(2, 2, figsize=(8,8))

# Loop over the axs and alpha values
for ax, alpha in zip(axs.flat, alphas):
    W = derivative_descent(df, alpha = alpha)
    ax.plot(domain, f(domain))
    ax.scatter(W, f(W), color = "magenta")
    ax.set_title(f"alpha = {alpha}, Iterations = {len(W)}", fontsize = 15)
```

## 2) Minimizing a function of several variables 

Given a function $f: \mathbb{R}^n \rightarrow \mathbb{R}$ of several variables $W =(w_1, w_2, \ldots, w_d)$, we want to find the minimum

$$
\min_{w_1, w_2, \ldots, w_d} f(w_1, w_2, \ldots, w_d)
$$

as well as the particular point $W=(w_1, w_2, \ldots, w_d)$ where this minimum is achieved. 

Again, we start with a random choice of $W$ and we use the gradient descent formula  

$$
W_{new} = W_{old} - \alpha \nabla f(W_{old}) 
$$

where $\nabla f$ is the gradient.

---

Consider a function $f(w_0, w_1) = w_0^2 + w_1^2 + 1$ which has an obvious minimum at $(w_0, w_1) = (0, 0)$. 

```python
#from mpl_toolkits import mplot3d

def f(w_0, w_1):
    return w_0 ** 2 + w_1 ** 2 + 1

x = np.linspace(-6, 6, 300)
y = np.linspace(-6, 6, 300)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize = (8, 8))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('f')

# This is to see the bottom of the function from a certain point
ax.view_init(60, 35);
```

```python
def df(w):
    grad = [2*w[0], 2*w[1]]
    return np.array(grad)

def gradient_descent(derivative, W, alpha = 0.8, max_iter = 1_000):
    i = 0
    W_hist = [W]
    while i < max_iter:
        W = W - alpha*df(W)
        i += 1
        W_hist.append(W)

    return W_hist

W = gradient_descent(df, np.array([5.0, -5.0]), alpha = 0.1, max_iter=500)

x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)


fig = plt.figure(figsize = (8, 8))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
for i in range(len(W)):
    ax.scatter3D([W[i][0]], 
                 [W[i][1]], 
                 [f(W[i][0], W[i][1])], color = "blue")
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('f')
ax.view_init(60, 35)
plt.show()
```
